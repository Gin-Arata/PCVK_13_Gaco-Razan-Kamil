```python
import matplotlib.pylab as plt
import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow
from PIL import Image, ImageDraw
```

```python
# @title Read image with opencv

image = cv.imread('KTP.jpeg')
cv2_imshow(image)
```

```python
# @title Inisialisasi tinggi dan lebar

tinggi = image.shape[0]
lebar = image.shape[1]
print(f'tinggi: {tinggi}, lebar: {lebar}')
```

```python
# @title Menggambar rectangle pada KTP

# Melakukan konversi dari openCV menjadi PIL Image
pil_image = Image.fromarray(image)

# Membuat variabel draw sebagai objek
draw = ImageDraw.Draw(pil_image)

# inisialisasi koordinat pada KTP untuk menggambar rectangle
x1, y1 = int(lebar/2) - 210, int(tinggi/2) + 40
x2, y2 = int(lebar/2) - 140, int(tinggi/2) + 55

x3, y3 = int(lebar/2) - 210, int(tinggi/2) + 60
x4, y4 = int(lebar/2) - 140, int(tinggi/2) + 75

# Menggambar rectangle pada agama dan status kawin
draw.rectangle([x1, y1, x2, y2], fill="black")
draw.rectangle([x3, y3, x4, y4], fill="white")

# Mengkonversi array PIL Image agar bisa dibuka menggunakan opencv dan plt
image_final_cv = np.array(pil_image)
image_final_plt = np.array(pil_image)

# Menampilkan gambar
cv2_imshow(image_final_cv)

# Melakukan konversi rgb
image_final_plt = cv.cvtColor(image_final_plt, cv.COLOR_BGR2RGB)
plt.imshow(image_final_plt)
```
