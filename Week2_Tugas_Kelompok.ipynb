# Tugas Kelompok

Tutup tiap bagian tertentu dari kartu identitas tersebut menggunakan fungsi yang telah
Anda pelajari. Kreasikan terkait dengan warna dan ukuran dari bentuk-bentuknya.
- Kelompok 1 tutup bagian nomor NIK dan nama.
- Kelompok 2 tutup bagian TTL dan Jenis Kelamin.
- Kelompok 3 tutup bagian Gol. Darah dan TTD.
- Kelompok 4 tutup bagian Alamat dan Kecamatan.
- Kelompok 5 tutup bagian Agama dan Status Kawin. Kelompok 6 tutup Pekerjaan dan
kewarganegaraan.
- Kelompok 7 tutup bagian berlaku hingga dan foto.
- Kelompok 8 tutup bagian kabupaten dan tanggal dikeluarkan KTP yang ada dibagian
bawah foto.
- Kelompok 9 tutup bagian nama propinsi dan kabupaten di bagian paling atas KTP.
- Kelompok 10 tutup bagian RT/RW dan Kel/desa. Tutup menggunakan 2 kotak warna
yang berbeda.

```python
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
from google.colab import drive
```

```python
# @title Melakukan mounting file dari drive

drive.mount('/content/drive')
```

```python
# @title Melakukan pengambilan gambar dari drive

img_ktp = cv.imread("/content/drive/MyDrive/PCVK_Minggu2/KTP.jpeg")

img_rgb = cv.cvtColor(img_ktp, cv.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
# @title Membuat vertices untuk menutup informasi

# Membuat vertices
vertices_agama = np.array([[190, 280], [250, 280], [250, 295], [190, 295]], dtype=np.int32)
vertices_kawin = np.array([[190, 300], [265, 300], [265, 325], [190, 325]], dtype=np.int32)

# Melakukan reshape vertices
vertices_agama = vertices_agama.reshape((-1, 1, 2))
vertices_kawin = vertices_kawin.reshape((-1, 1, 2))
```

```python
# @title Melakukan Penutupan Identitas Agama dan Kawin

# Menambahkan polylines pada agama dan status kawin
cv.fillPoly(img_rgb, pts=[vertices_agama], color=(255, 255, 255))
cv.fillPoly(img_rgb, pts=[vertices_kawin], color=(0, 0, 0))

# Menampilkan gambar yang sudah dilakukan penutupan identitas
plt.imshow(img_rgb)
```
