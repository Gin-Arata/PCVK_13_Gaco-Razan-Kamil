## Implementasi template matching tanpa library

```python
# @title Read image

img_baseBahrain = cv.imread('/content/drive/MyDrive/PCVK_Minggu2/Minggu 2/Images/Object Detection/bahrain.jpg')
img_bahrainTemplate = cv.imread('/content/drive/MyDrive/PCVK_Minggu2/Minggu 2/Images/Object Detection/bahrain-template.jpg')

# Convert warna rgb
img_baseBahrain = cv.cvtColor(img_baseBahrain, cv.COLOR_BGR2RGB)
img_bahrainTemplate = cv.cvtColor(img_bahrainTemplate, cv.COLOR_BGR2RGB)
```

```python
# @title Function template matching

# Fungsi template matching
def template_matching(image, template, threshold=0.5, overlap_thresh=0.5):
  h_img, w_img = image.shape[:2]
  h_temp, w_temp = template.shape[:2]

  # Mencari kecocokan berdasarkan korelasi
  matches = [
      (x, y, np.corrcoef(image[y:y+h_temp, x:x+w_temp].flatten(), template.flatten())[0, 1])
      for y in range(h_img - h_temp + 1) for x in range(w_img - w_temp + 1)
      if np.corrcoef(image[y:y+h_temp, x:x+w_temp].flatten(), template.flatten())[0, 1] >= threshold
  ]

  # Mengurutkan kecocokan berdasarkan nilai korelasi tertinggi
  matches.sort(key=lambda x: x[2], reverse=True)

  # Melakukan filter terhadap nilai korelasi yang saling tumpang tindih
  final_matches, suppressed = [], [False] * len(matches)
  for i, (x1, y1, _) in enumerate(matches):
      if suppressed[i]:
          continue
      final_matches.append((x1, y1))
      for j, (x2, y2, _) in enumerate(matches[i+1:], start=i+1):
        if not suppressed[j]:
          # Menghitung area yang bertindihan
          overlap_x = max(0, min(x1 + w_temp, x2 + w_temp) - max(x1, x2))
          overlap_y = max(0, min(y1 + h_temp, y2 + h_temp) - max(y1, y2))
          overlap_area = overlap_x * overlap_y
          if overlap_area / float(2 * w_temp * h_temp - overlap_area) > overlap_thresh:
              suppressed[j] = True

  return final_matches

# Funtion gambar rectangle
def draw_matches(image, matches, template_size):
  img_pil = Image.fromarray(image)
  draw = ImageDraw.Draw(img_pil)
  for (x, y) in matches:
      draw.rectangle([x, y, x + template_size[1], y + template_size[0]], outline="green", width=2)
  return img_pil
```

```python
# @title Melakukan template matching

input_image = cv.imread('/content/drive/MyDrive/PCVK_Minggu2/Minggu 2/Images/Object Detection/bahrain.jpg')
template_image = cv.imread('/content/drive/MyDrive/PCVK_Minggu2/Minggu 2/Images/Object Detection/bahrain-template.jpg')

matches = template_matching(input_image, template_image, threshold=0.45)
result_image = draw_matches(input_image, matches, template_image.shape[:2])

plt.imshow(result_image)
plt.axis('off')
plt.show()
```

## Implementasi edge detection dengan openCV

```python
# @title Read Image

img_carPark = cv.imread('/content/drive/MyDrive/PCVK_Minggu2/Minggu 2/Images/Object Detection/car-park.jpg')

# Convert warna rgb
img_carPark = cv.cvtColor(img_carPark, cv.COLOR_BGR2RGB)
```

```python
# @title Implementasi Sobel Edge

img_sobelDetect = img_carPark.copy()

# Convert color to gray
img_sobelDetect = cv.cvtColor(img_sobelDetect, cv.COLOR_RGB2GRAY)

# Melakukan sobel detection
sobel_x = cv.Sobel(img_sobelDetect, cv.CV_64F, 1, 0, ksize=5)  # Deteksi tepi horizontal
sobel_y = cv.Sobel(img_sobelDetect, cv.CV_64F, 0, 1, ksize=5)  # Deteksi tepi vertikal
sobel_combined = cv.magnitude(sobel_x, sobel_y)

# Melakukan perhitungan gradient
# abs_sobelx = cv.convertScaleAbs(sobelx)
# abs_sobely = cv.convertScaleAbs(sobely)
# grad_sobel = cv.addWeighted(abs_sobelx, 0.5, abs_sobely, 0.5, 0)

# Menampilkan image
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].imshow(img_sobelDetect, cmap='gray')
axes[0].set_title('Original Image')

axes[1].imshow(sobel_combined, cmap='gray')
axes[1].set_title('Sobel Edge Detection')
```

```python
# @title Implementasi canny detect

img_cannyDetect = img_carPark.copy()

# Convert color to gray
img_cannyDetect = cv.cvtColor(img_cannyDetect, cv.COLOR_RGB2GRAY)

# Melakukan canny detection
canny_edges = cv.Canny(img_cannyDetect, threshold1= 100, threshold2=200)

# Menampilkan image
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].imshow(img_cannyDetect, cmap='gray')
axes[0].set_title('Original Image')

axes[1].imshow(canny_edges, cmap='gray')
axes[1].set_title('Canny Edge Detection')
```

```python
# @title Implementasi laplacian detect

img_laplacianDetect = img_carPark.copy()

# Convert color to gray
img_laplacianDetect = cv.cvtColor(img_laplacianDetect, cv.COLOR_RGB2GRAY)

# Melakukan laplacian
laplacian = cv.Laplacian(img_laplacianDetect, cv.CV_64F)

# Hitung absolute value dan convert ke 8-bit
laplacian = cv.convertScaleAbs(laplacian)

# Menampilkan image
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].imshow(img_laplacianDetect, cmap='gray')
axes[0].set_title('Original Image')

axes[1].imshow(laplacian, cmap='gray')
axes[1].set_title('Laplacian Edge Detection')
```

## Implementasi Corner Detection

```python
# @title Read Image

img_corner = cv.imread('/content/drive/MyDrive/PCVK_Minggu2/Minggu 2/Images/Object Detection/chess-board.jpg')

# Conver color to gray
img_corner_gray = cv.cvtColor(img_corner, cv.COLOR_BGR2GRAY)
```

```python
# @title Harris corner detection

# Copy image
img_corner_harris = img_corner_gray.copy()

harris_corner = cv.cornerHarris(
    img_corner_harris,
    blockSize=2,
    ksize=3,
    k=0.04
  )

harris_corner = cv.dilate(harris_corner, None)

img_harris = img_corner.copy()
img_harris = cv.cvtColor(img_harris, cv.COLOR_BGR2RGB)

# Menggunakan np.where untuk mendapatkan koordinat piksel
y, x = np.where(harris_corner > 0.01 * harris_corner.max())
for i in range(len(x)):
    cv.circle(img_harris, (x[i], y[i]), 2, (0, 0, 255), -1)

# Tampilkan hasil
plt.imshow(img_harris)
plt.title('Harris Corner Detection')
plt.show()
```

```python
# @title Shi Tomashi Corner

img_shiTomashi = img_corner_gray.copy()

shi_tomashi_corner = cv.goodFeaturesToTrack(
    img_shiTomashi,
    maxCorners=100,
    qualityLevel=0.01,
    minDistance=10
  )

corners = np.int32(shi_tomashi_corner)
img_chessBoard = img_corner.copy()
img_chessBoard = cv.cvtColor(img_chessBoard, cv.COLOR_BGR2RGB)

for i in corners:
    x, y = i.ravel()
    cv.circle(img_chessBoard, (x, y), 5, (255, 0, 0), -1)

# Tampilkan hasil
plt.imshow(img_chessBoard)
plt.title('Harris Corner Detection')
plt.show()
```

## Implementasi hough transform

```python
# @title Read Img

img_hough = cv.imread('/content/drive/MyDrive/PCVK_Minggu2/Minggu 2/Images/Object Detection/sudoku.jpg')

# COnver color to gray
img_houghGray = cv.cvtColor(img_hough, cv.COLOR_BGR2GRAY)
```

```python
# @title Melakukan pendeteksian tepi menggunakan canny

edges = cv.Canny(img_houghGray, threshold1=50, threshold2=150)
```

```python
# @title Melakukan pendeteksian hough transform

line = cv.HoughLinesP(edges, 1, np.pi / 180, threshold=100, minLineLength=50, maxLineGap=10)

# Membuat garis dari hasil hough
image_with_lines = img_hough.copy()
if line is not None:
    for line in line:
        x1, y1, x2, y2 = line[0]
        cv.line(image_with_lines, (x1, y1), (x2, y2), (0, 0, 255), 2)

# Menampilkan img

plt.imshow(image_with_lines)
plt.title('Hough Transform')
plt.axis('off')
```

```python
# @title Melakukan threshold terhadap KTP

ktp_img = cv.imread('/content/drive/MyDrive/PCVK_Minggu2/ktp3.png')

# Melakukan blurring pada image
blurred = cv.GaussianBlur(ktp_img, (5, 5), 0.5)

# Convert ke grayscale
gray = cv.cvtColor(blurred, cv.COLOR_BGR2GRAY)

# Melakukan thresholding menggunakan binary dan otsu
ret, img_threshold = cv.threshold(gray.copy(), 0, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)


plt.figure(figsize=(10, 10))
plt.imshow(cv.cvtColor(img_threshold, cv.COLOR_BGR2RGB)), plt.axis('off')
plt.show()
```
